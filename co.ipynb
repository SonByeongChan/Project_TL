# '개노가다'로 바운딩 안 맞는 데이터 삭제 후 남은 파일의 갯수

```python
import os

# 이미지가 저장된 폴더 경로
folder_path = r'E:\AI_KDT7\12.Transfer_learning\mini\xray_img\IMG'

# 이미지 확장자 리스트 (필요에 따라 확장자 추가 가능)
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.gif']

# 폴더 내 모든 파일 중 이미지 파일만 필터링
image_files = [f for f in os.listdir(folder_path)
               if os.path.isfile(os.path.join(folder_path, f)) and
               os.path.splitext(f)[1].lower() in image_extensions]

print(f"폴더 내 이미지 파일 개수: {len(image_files)}")
```

```python
import os
from lxml import etree

# 1) 이미지 파일명 추출 함수
def get_image_files(folder_path, extensions=None):
    if extensions is None:
        extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']
    files = [f for f in os.listdir(folder_path)
             if os.path.isfile(os.path.join(folder_path, f))
             and os.path.splitext(f)[1].lower() in extensions]
    return files

# 2) XML 파일 매칭 함수
def match_xml_files(image_files, xml_folder):
    matched_xml_files = []
    for img_name in image_files:
        base_name = os.path.splitext(img_name)[0]
        xml_path = os.path.join(xml_folder, base_name + '.xml')
        if os.path.isfile(xml_path):
            matched_xml_files.append(xml_path)
            print(f"Found XML: {xml_path}")
        else:
            print(f"XML not found for image: {img_name}")
    return matched_xml_files

# 3) VOC XML 파싱 함수
def parse_voc_xml(xml_path):
    tree = etree.parse(xml_path)
    root = tree.getroot()

    filename = root.find('filename').text

    objects = []
    for obj in root.findall('object'):
        cls = obj.find('name').text
        bbox = obj.find('bndbox')
        xmin = int(bbox.find('xmin').text)
        ymin = int(bbox.find('ymin').text)
        xmax = int(bbox.find('xmax').text)
        ymax = int(bbox.find('ymax').text)
        objects.append({'class': cls, 'bbox': (xmin, ymin, xmax, ymax)})

    return filename, objects

# --- 사용 예시 ---

# 이미지 폴더 경로
image_folder = r'E:\AI_KDT7\12.Transfer_learning\mini\xray_img\IMG'

# XML 폴더 경로
xml_folder = r'E:\AI_KDT7\12.Transfer_learning\mini\xray_img\Annotation\Train\Pascal\Astrophysics'

# 1) 이미지 파일명 리스트 불러오기
image_files = get_image_files(image_folder)
print(f"총 이미지 파일 개수: {len(image_files)}")

# 2) XML 파일 매칭
matched_xml_files = match_xml_files(image_files, xml_folder)
print(f"총 매칭된 XML 파일 개수: {len(matched_xml_files)}")

# 3) XML 파일 파싱 예시
for xml_file in matched_xml_files:
    filename, objects = parse_voc_xml(xml_file)
    print(f"파일명: {filename}")
    for obj in objects:
        print(f"클래스: {obj['class']}, 바운딩박스: {obj['bbox']}")
    print('---')
    # 예시로 첫 파일만 처리하려면 break 해제
    break
```

```python
import os
import shutil

# 기준 폴더 (이름 기준으로 사용할 파일들이 있는 폴더)
base_folder = r'E:\AI_KDT7\12.Transfer_learning\mini\xray_img\IMG'

# 원본 이미지가 저장된 폴더 (복사 대상 파일들이 있는 폴더)
source_folder = r'E:\AI_KDT7\12.Transfer_learning\mini\xray_img\Astrophysics\[Astro]ZippoOil\ZippoOil\Single_Default'

# 복사할 대상 폴더
target_folder = r'E:\AI_KDT7\12.Transfer_learning\mini\project_root\train_image'

# 확장자 허용 리스트 (필요하면 조정 가능)
extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.tiff']

# 대상 폴더가 없으면 생성
os.makedirs(target_folder, exist_ok=True)

# 1) base_folder의 확장자 제외한 파일명 집합 생성
base_files = os.listdir(base_folder)
base_names = set(os.path.splitext(f)[0] for f in base_files if os.path.splitext(f)[1].lower() in extensions)

# 2) source_folder의 파일 중 base_names에 해당하는 이름을 가진 파일들을 찾아 복사
source_files = os.listdir(source_folder)
copied_count = 0

for f in source_files:
    name, ext = os.path.splitext(f)
    if ext.lower() in extensions and name in base_names:
        src_path = os.path.join(source_folder, f)
        dst_path = os.path.join(target_folder, f)
        shutil.copy2(src_path, dst_path)
        print(f"복사 완료: {f}")
        copied_count += 1

print(f"총 복사된 파일 수: {copied_count}")
```
